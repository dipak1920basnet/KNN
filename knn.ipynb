```python
import numpy as np 
from collections import Counter
```

```python
# Calculate the eculidean Distance from the points 
def euclidean_distance(point1, point2):
    distance = np.sum((np.array(point1) - np.array(point2))**2)
    return np.sqrt(distance)
```

```python
def knn_predict(training_data, training_labels, test_point,k):
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist,training_labels[i]))
    distances.sort(key=lambda x:x[0])
    k_nearest_lables = [labels for _,labels in distances[:k]]

    return Counter(k_nearest_lables).most_common(1)[0][0]
```

```python
# training Data labels and testing points 
training_data = [[1, 2], [2, 3], [3, 4], [6, 7], [7, 8]]
training_labels = ["A","A","A","B","B"]
test_point = [4,5]
k = 3
```

```python
predictions = knn_predict(training_data, training_labels, test_point, k)
```
